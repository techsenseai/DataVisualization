# Visualización de Datos en Python: Gráficos y Documentación Completa

La visualización es clave para explorar y entender mejor los datos. Aquí te muestro cómo usar las bibliotecas Matplotlib, Seaborn y Plotly para crear una variedad de gráficos, desde básicos hasta avanzados.

## 1. Matplotlib: Visualización Básica y Personalización Avanzada
Matplotlib es la biblioteca de visualización más versátil en Python y permite crear una amplia gama de gráficos personalizados.

Documentación de Matplotlib
Página de la Documentación Oficial de Matplotlib
Gráfico de Área
Los gráficos de área son ideales para representar la acumulación de cantidades a lo largo del tiempo.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
meses = ['Enero', 'Febrero', 'Marzo', 'Abril']
ventas = [2500, 2700, 3000, 2800]

# Gráfico de área
plt.fill_between(meses, ventas, color="skyblue", alpha=0.4)
plt.plot(meses, ventas, color="Slateblue", alpha=0.6)
plt.title('Ventas Mensuales (Área)')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.show()
```

## Gráfico de Radar
Un gráfico de radar es útil para comparar múltiples variables categóricas.

```python
import numpy as np

# Datos de ejemplo
categorias = ['Cualidad A', 'Cualidad B', 'Cualidad C', 'Cualidad D']
valores = [4, 3, 2, 5]

# Configuración de gráfico de radar
valores += valores[:1]
categorias += categorias[:1]
angulo = np.linspace(0, 2 * np.pi, len(categorias))

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angulo, valores, color='b', linewidth=2)
ax.fill(angulo, valores, color='b', alpha=0.25)
plt.title('Comparación de Cualidades')
plt.show()
```

## Gráfico de Violin
El gráfico de violín combina características de box plot y kernel density plot, mostrando la distribución y densidad de los datos.

```python
import seaborn as sns
import pandas as pd

# Datos de ejemplo
data = pd.DataFrame({'Categoría': np.random.choice(['A', 'B', 'C'], 100),
                     'Valor': np.random.normal(0, 1, 100)})

# Gráfico de violín
sns.violinplot(x='Categoría', y='Valor', data=data)
plt.title('Distribución de Valores por Categoría')
plt.show()
```

## 2. Seaborn: Visualización Estadística y Análisis
Seaborn facilita la creación de gráficos complejos y estilizados que incluyen elementos estadísticos.

### Gráfico de KDE (Kernel Density Estimation)
El KDE es una visualización de la densidad de una variable continua y es útil para analizar la distribución de los datos.

```python
# Datos de ejemplo
edades = np.random.normal(30, 5, 100)

# Gráfico de KDE
sns.kdeplot(edades, shade=True, color='r')
plt.title('Distribución de Densidad de Edad')
plt.xlabel('Edad')
plt.show()
```

### Gráfico de Pares (Pairplot)
El gráfico de pares muestra relaciones entre todas las variables numéricas de un DataFrame y es muy útil para el análisis exploratorio.

```python
# Datos de ejemplo
df = pd.DataFrame({
    'Ingresos': np.random.normal(50000, 10000, 100),
    'Gastos': np.random.normal(30000, 5000, 100),
    'Edad': np.random.randint(20, 50, 100)
})

# Pairplot
sns.pairplot(df)
plt.show()
```

### Swarm Plot
El swarm plot es ideal para mostrar la distribución de puntos en una categoría sin que se sobrepongan.

```python
# Swarm Plot
sns.swarmplot(x='Categoría', y='Valor', data=data)
plt.title('Swarm Plot de Valores por Categoría')
plt.show()
```

## 3. Plotly: Visualización Interactiva
Plotly permite crear gráficos interactivos y personalizables. Es ideal para mostrar visualizaciones que los usuarios puedan explorar.

### Gráfico de Rosca (Donut Chart)
Similar a un gráfico de pastel, pero con el centro vacío para ahorrar espacio visual.

```python
import plotly.express as px

# Datos de ejemplo
df_donut = pd.DataFrame({
    'Categoría': ['A', 'B', 'C'],
    'Valor': [45, 25, 30]
})

# Gráfico de rosca
fig = px.pie(df_donut, values='Valor', names='Categoría', hole=0.3, title='Distribución de Categorías')
fig.show()
```

### Gráfico de Dispersión 3D
Un gráfico de dispersión en 3D ayuda a visualizar relaciones entre tres variables.

```python
# Datos de ejemplo
df_3d = pd.DataFrame({
    'Ingresos': np.random.normal(50000, 10000, 100),
    'Gastos': np.random.normal(30000, 5000, 100),
    'Edad': np.random.randint(20, 50, 100)
})

# Gráfico de dispersión 3D
fig = px.scatter_3d(df_3d, x='Ingresos', y='Gastos', z='Edad', color='Edad', title='Ingresos, Gastos y Edad')
fig.show()
```

## Gráfico de Mapas Coropletos
Plotly permite crear mapas coropletos para visualizar datos geográficos.

```python
# Datos de ejemplo
df_map = pd.DataFrame({
    'País': ['México', 'Argentina', 'Brasil'],
    'Población': [126200000, 44940000, 212600000]
})

# Gráfico de mapa coroplético
fig = px.choropleth(df_map, locations='País', locationmode='country names',
                    color='Población', title='Población por País en América Latina',
                    color_continuous_scale='Viridis')
fig.show()
```

# Análisis de Series de Tiempo con Pandas
El análisis de series de tiempo es esencial cuando trabajas con datos ordenados cronológicamente, como ventas diarias, precios de acciones o datos climáticos. Pandas ofrece varias herramientas para manipular y analizar estos datos.

## Creación y Manipulación de Series de Tiempo
Para trabajar con series de tiempo en Pandas, el primer paso es asegurarse de que las fechas están correctamente formateadas. Puedes usar pd.to_datetime() para convertir una columna de fechas en un formato que Pandas pueda reconocer como serie de tiempo.

```python
import pandas as pd
import numpy as np

# Crear un DataFrame de ejemplo con datos diarios
fechas = pd.date_range(start='2023-01-01', periods=100, freq='D')
datos = np.random.normal(100, 10, len(fechas))

df = pd.DataFrame({'Fecha': fechas, 'Ventas': datos})
df['Fecha'] = pd.to_datetime(df['Fecha'])
df.set_index('Fecha', inplace=True)
print(df.head())
```

## Visualización de Series de Tiempo
Con los datos de series de tiempo organizados, podemos graficarlos para detectar patrones y tendencias.

```python
import matplotlib.pyplot as plt

# Gráfico de la serie de tiempo
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Ventas'], color='blue')
plt.title('Ventas Diarias')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.show()
```

## Resampleo y Agregación
Pandas facilita el resampleo de datos, lo cual es útil para agrupar series de tiempo en intervalos más amplios, como semanas o meses.

```python
# Resampleo mensual
ventas_mensuales = df['Ventas'].resample('M').mean()
print(ventas_mensuales)

# Gráfico de la serie de tiempo mensual
ventas_mensuales.plot(title='Ventas Mensuales', color='green')
plt.xlabel('Fecha')
plt.ylabel('Ventas Promedio')
plt.show()
```

## Detección de Estacionalidad y Tendencias
Pandas también puede combinarse con otras bibliotecas, como statsmodels, para detectar estacionalidad y tendencias en series de tiempo.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Descomposición de la serie de tiempo
descomposicion = seasonal_decompose(df['Ventas'], model='additive', period=30)
descomposicion.plot()
plt.show()
```

## Pronósticos Básicos
Una vez que se comprenden las tendencias y patrones, Pandas también puede trabajar en conjunto con bibliotecas de machine learning para realizar pronósticos. Para empezar, puedes realizar un pronóstico básico utilizando el último valor conocido o un promedio móvil.

```python
# Pronóstico simple con promedio móvil
df['Pronostico'] = df['Ventas'].rolling(window=5).mean()
df[['Ventas', 'Pronostico']].plot(figsize=(10, 6))
plt.title('Pronóstico con Promedio Móvil')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.show()
```

Estas herramientas y métodos son fundamentales para comprender el comportamiento de los datos a lo largo del tiempo y realizar análisis predictivo. Con Pandas, el análisis de series de tiempo es sencillo y flexible, permitiéndote obtener información clave para la toma de decisiones.
